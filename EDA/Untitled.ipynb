```python
# Import the libraries
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

data = pd.read_csv('../Datasets/Cleaned_data.csv', index_col = [0])
data = data.drop('Assigned.ID', axis = 1)
```

```python
# Binary encoding for specific columns
binary_encoding = ['SpcHsgGoals', 'Self', 'First', 'Geog']
data[binary_encoding] = data[binary_encoding].replace([2], 0)

data['CoEth'] = data['CoEth'].replace([5], 3)
data['CoRace'] = data['CoRace'].replace([8], 7)
data['CoGender'] = data['CoGender'].replace([4], 3)
data['CoAge'] = data['CoAge'].replace([98], 99)
```
