```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

data = pd.read_csv('../Datasets/Cleaned_data.csv', index_col = [0])
```

```python
numerical_cols = ['BoAge', 'MinPer', 'Tractrat', 'Income',\
                 'IncRat', 'UPB', 'LTV', 'Term', 'Rate', 'Front','Back', 'PMI']

categorical_cols = ['BoRace', 'BoGender',\
                  'PropType', 'AcqTyp', 'FedGuar']

ordinal_cols = ['BoCreditScor', 'NumBor']
```

```python
df = pd.concat([data[numerical_cols],data[ordinal_cols]], axis = 1)
```

```python
corr = df.corr('pearson')
```

```python
import seaborn as sns
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```
